```python
# 1. 중첩 함수 (Nested Function)
# 문제 1-1. 평균 구하기 리스트 하나를 인자로 받는 함수 get_stats(numbers)를 만드세요.

# 함수 내부에 리스트의 합계를 구하는 sum_list()라는 중첩 함수를 만드세요.

# get_stats는 sum_list를 이용해 평균값을 계산하여 반환해야 합니다.

# Python
# # 사용 예시
# result = get_stats([10, 20, 30, 40])
# print(result) 

# 기대 출력
# 25.0
```

```python
def get_stats(numbers):
    lst = numbers
    def sum_list():
        return sum(numbers)

    total = sum_list()
    average = total / len(numbers)

    return average

result = get_stats([10, 20, 30, 40])
print(result)
```

```python
# 문제 1-2. 인사말 생성기 greeting(name, mode) 함수를 만드세요.

# mode가 0이면 정중한 인사("안녕하십니까, {name}님."), 1이면 친근한 인사("안녕, {name}!")를 반환해야 합니다.

# 함수 내부에 각각의 메시지 포맷을 처리하는 중첩 함수 2개를 정의해서 사용하세요. (예: formal(), casual())

# Python
# # 사용 예시
# print(greeting("민수", 0))
# print(greeting("민수", 1))

# 기대 출력
# 안녕하십니까, 민수님.
# 안녕, 민수!
```

```python
def greeting(name, mode):    
    def formal():
        return f"안녕하십니까, {name}님."
    
    def casual():
        return f"안녕, {name}!"
    
    if mode == 0:
        msg = formal()
    else:
        msg = casual()
    return msg
        
print(greeting("민수", 0))    
print(greeting("민수", 1))
```

```python
# 2. 클로저 (Closure)
# 문제 2-1. N배수 생성 공장 make_multiplier(n) 함수를 작성하세요.

# 이 함수는 숫자 하나를 입력받아, 처음에 설정한 n을 곱해주는 함수(클로저)를 반환해야 합니다.

# Python
# # 사용 예시
# doubler = make_multiplier(2)  # 2배 만드는 함수 생성
# tripler = make_multiplier(3)  # 3배 만드는 함수 생성

# print(doubler(10))
# print(tripler(10))

# 기대 출력
# 20
# 30
```

```python
def make_multiplier(n):

    def multiplier(x):
        return n * x

    return multiplier

doubler = make_multiplier(2)
tripler = make_multiplier(3)

print(doubler(10))
print(tripler(10))
```

```python
# 문제 2-2. 누적 저금통 piggy_bank() 함수를 작성하세요.

# 이 함수는 초기 잔액을 0으로 가지고 있고, 호출할 때마다 금액을 입력받아 누적된 총액을 반환하는 함수(클로저)를 반환해야 합니다.

# nonlocal 키워드를 사용해야 합니다.

# Python
# # 사용 예시
# my_bank = piggy_bank()

# print(my_bank(1000)) # 1000원 저금
# print(my_bank(500))  # 500원 저금
# print(my_bank(2000)) # 2000원 저금

# 기대 출력
# 1000
# 1500
# 3500
```

```python
def piggy_bank():
    money = 0
    def deposit(m):
        nonlocal money
        money += m
        return money
    return deposit

my_bank = piggy_bank()

print(my_bank(1000))
print(my_bank(500))
print(my_bank(2000))
```

```python
# 3. 데코레이터 (Decorator)
# 문제 3-1. 함수 실행 알리미 @start_end_printer 라는 데코레이터를 만드세요.

# 어떤 함수가 실행되기 전에 "== Start ==", 실행이 끝난 후에 "== End =="를 출력해야 합니다.

# Python
# @start_end_printer
# def my_func():
#     print("함수 실행 중...")

# # 사용 예시
# my_func()

# 기대 출력
# == Start ==
# 함수 실행 중...
# == End ==
```

```python
def start_end_printer(func):
    def wrapper():
        print("== Start ==")
        func()
        print("== End ==")
    return wrapper

@start_end_printer
def my_func():
    print("함수 실행 중...")

my_func()
```

```python
# 문제 3-2. HTML 태그 씌우기 @add_html_tag 데코레이터를 만드세요.

# 함수가 반환(Return)하는 문자열 앞뒤에 <div>와 </div> 태그를 붙여주는 데코레이터입니다.

# Python
# @add_html_tag
# def get_text():
#     return "Hello Python"

# 사용 예시
# print(get_text())

# 기대 출력
# <div>Hello Python</div>
```

```python
def add_html_tag(func):
    def wrapper():
        return "<div>" + func() + "</div>"
    return wrapper
@add_html_tag
def get_text():
    return "Hello Python"

print(get_text())
```
